```python
# ==== Dual Independent Plot Visualization ====
import numpy as np
import matplotlib.pyplot as plt

def plot_dual_gam(csv_path, xmatch, OmegaSq):
    try:
        # Load and process data
        data = np.genfromtxt(csv_path, delimiter=',', skip_header=1)
        x = data[:,0]
        Er = data[:,1]
        
        # Create two separate plots
        plt.figure(figsize=(16, 6))
        
        # Left plot (0 <= x <= xmatch)
        plt.subplot(1, 2, 1)  # 1 row, 2 columns, first plot
        left_mask = x <= xmatch
        plt.plot(x[left_mask], Er[left_mask], 
                color='#1f77b4', linewidth=2.5,
                label=f'Left Domain (0 ≤ x ≤ {xmatch:.2f})')
        plt.axvline(xmatch, color='r', linestyle=':', linewidth=1.5)
        plt.grid(alpha=0.3, linestyle=':')
        plt.xlabel('Normalized Radius', fontsize=12)
        plt.ylabel('Re($E_r$) Magnitude', fontsize=12)
        plt.title(f'Left GAM Profile (Ω² = {OmegaSq:.3f})', pad=15)
        plt.legend(loc='upper right')
        plt.xlim(0, xmatch*1.02)
        
        # Right plot (xmatch < x <= 1)
        plt.subplot(1, 2, 2)  # 1 row, 2 columns, second plot
        right_mask = x > xmatch
        plt.plot(x[right_mask], Er[right_mask], 
                color='#d62728', linewidth=2.5, linestyle='--',
                label=f'Right Domain ({xmatch:.2f} < x ≤ 1)')
        plt.axvline(xmatch, color='r', linestyle=':', linewidth=1.5)
        plt.grid(alpha=0.3, linestyle=':')
        plt.xlabel('Normalized Radius', fontsize=12)
        plt.title(f'Right GAM Profile (Ω² = {OmegaSq:.3f})', pad=15)
        plt.legend(loc='upper left')
        plt.xlim(xmatch*0.98, 1)
        
        # Global settings
        plt.suptitle(f'Global GAM Mode Structure Split at x={xmatch:.2f}', 
                    y=1.02, fontsize=14, weight='bold')
        plt.tight_layout()
        plt.show()
        
    except Exception as e:
        print(f"Visualization Error: {str(e)}")

# Execution example
plot_dual_gam("GAM_exact_data.csv", xmatch=0.8, OmegaSq=1.03)
```
